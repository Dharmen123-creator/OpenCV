```
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = np.zeros((400,400,3), dtype = 'uint8')
a = img1.copy()
rectangle = cv2.rectangle(a, (50,50), (350,350), (0,0,255), 8)

plt.imshow(rectangle)
plt.show()
```

```
b = img1.copy()

(x,y) = (int(img1.shape[1]/2), int(img1.shape[0]/2))

circle = cv2.circle(b, (x,y), (150), (0,0,255), 8)
plt.imshow(circle)
plt.show()
```

```
bitwiseXOR = cv2.bitwise_xor(rectangle,circle)
plt.imshow(bitwiseXOR)
plt.show()
```

```
bitwiseNOT = cv2.bitwise_not(rectangle,circle)
plt.imshow(bitwiseNOT)
plt.show()
```

```
bitwiseAND = cv2.bitwise_and(circle,rectangle)
plt.imshow(bitwiseAND)
plt.show()
```
