```
 from google.colab.patches import cv2_imshow
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Micky_Mouse.jpg',-1)
smallerimg = cv2.pyrDown(img)
largerimg = cv2.pyrUp(smallerimg)

cv2_imshow(img)
cv2_imshow(smallerimg)
cv2_imshow(largerimg)
cv2.destroyAllWindows()
```
